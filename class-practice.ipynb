```python
class MyAdd:
    def __init__(self,a):
        self.a = a
        
    def __call__(self,b):
        print(f'{self.a}+b = {self.a + b}')
    
m = MyAdd(2)


m(4)
```

```python
import numpy as np 

class Avg:
    def __init__(self):
        self.xs = []
        
    def __call__(self,x):
        self.xs.append(x)
        return np.mean(self.xs)

avg = Avg()
avg(10), avg(11), avg(12)
    
# xs =[]

# def avg(x):
#     xs.append(x)
#     return np.mean(xs)

# avg(10),avg(11)
```
